```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def plot_graph(path, len_df = 120):
    df = pd.read_csv(path, encoding='cp1251', sep = ';', decimal=',')
    df_plot = df.iloc[:len_df]

    fig, ax = plt.subplots(figsize = (15, 7))

    for i in df.columns[3:]:
        x = np.arange(0, len_df, 1)
        y = np.array(df_plot[i])

        ax.plot(x, y, label = i)

    ax.set(xlabel='Время (минуты с начала измерения)', ylabel='Температура (С)',
           title='График температуры термокосы в зависимости от времени экспозиции')
    ax.grid()
    ax.set_xlim([0, len_df])
    ax.legend()
    name_of_graph = path[:-4] + '.png'
    fig.savefig(name_of_graph)
    plt.show()
    
```

```python
len_df = 120 # длительность записи в минутах (при частоте записи в одну минуту!!!)
path = '2022 осень/ОМР/38.csv'
plot_graph(path = path, len_df)
```
